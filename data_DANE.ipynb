```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns=None
```

```python
data_dane = pd.read_csv('../data/TerriData_utf8.txt', sep='|')
```

```python
data_dane.head()
```

```python
data_dane.info()
```

```python
data_dane['Dato Numérico'] = data_dane['Dato Numérico'].str.replace('.','')
data_dane['Dato Numérico'] = data_dane['Dato Numérico'].str.replace(',','.')
data_dane['Dato Numérico'] = data_dane['Dato Numérico'].astype('float64')
```

```python
dict_ = {
    'á' : 'a',
    'é' : 'e',
    'í' : 'i',
    'ó' : 'o',
    'ú' : 'u',
    'ü' : 'u',
    'ñ' : 'ni'
}

def remove_tildes(ser):
    new_list = []
    for element in list(ser):
        new_item = []
        new_item.append([dict_.get(char, char) for char in element])
        new_item = ''.join(new_item[0])
        new_list.append(new_item)
    return new_list
```

```python
data_dane.columns = remove_tildes(data_dane.columns.str.lower().str.replace(' ', '_'))
```

```python
for col in data_dane.columns:
    try:
        if data_dane[col].dtype == 'object':
            data_dane[col] = data_dane[col].str.replace(',', ';')
            data_dane[col] = remove_tildes(data_dane[col].str.lower())
            data_dane[col] = data_dane[col].str.replace(' ', '_')
    except(TypeError):
        pass
```

```python
data_dane['unidad_de_medida'] = data_dane['unidad_de_medida'].fillna('na')
data_dane['unidad_de_medida'] = remove_tildes(data_dane['unidad_de_medida'].str.lower().str.replace(' ', '_').str.replace('.',''))
```

```python
dims = list(data_dane['dimension'].unique())
new_dfs = []
for dim in dims:
    new_dfs.append(data_dane[data_dane['dimension']==dim])
```
